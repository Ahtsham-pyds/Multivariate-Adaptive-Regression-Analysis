```python
a=data.columns
data.drop(a[0],axis=1,inplace=True)
```

### Finding most correlated columns


```python
corre=[]
for i in range(len(data.columns)):
    a=data[[data.columns[i],'y1']].corr().iloc[0,1]
    if a>0.12:
        #print(semidata[[semidata.columns[i],'y1']].corr())
        datagramee=data[[data.columns[i],'y1']].corr().reset_index(drop=True).columns[0]
        corre.append(datagramee)
        #corre=corre.append(datagramee,sort=True)
     
```

```python
correleted_data=data[corre]
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt 
import numpy as np 
from sklearn import datasets, linear_model, metrics 
import pandas as pd
```

```python
X=correleted_data.drop(columns=['y1'])
y=correleted_data.y1
```

### Scaling Data

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X=pd.DataFrame(scaler.fit_transform(X),columns=X.columns)
```

### Linear Regression

```python
# variance_score=[]
# for i in range(42):    
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=26)
reg = linear_model.LinearRegression() 
reg.fit(X_train, y_train) 

# #     # regression coefficients 
print('Coefficients: \n', reg.coef_)

# #     # variance score: 1 means perfect prediction 
print('Variance score: {}'.format(reg.score(X_test, y_test))) 
#   variance_score.append(reg.score(X_test, y_test))

# # #     # plot for residual error 

# # #     ## setting plot style 
plt.style.use('fivethirtyeight') 

# # #     ## plotting residual errors in training data 

plt.scatter(reg.predict(X_train),reg.predict(X_train) - y_train, color = "green", s = 5) 

# # #     ## plotting residual errors in test data 
plt.scatter(reg.predict(X_test), reg.predict(X_test) - y_test, 
           color = "blue", s = 5) 

# # #     ## plotting line for zero residual error 
plt.hlines(y = 0, xmin = 0.06, xmax = .11, linewidth = 1) 

# # #     ## plotting legend 
plt.legend(loc = 'upper right') 

# # #     ## plot title 
plt.title("Residual errors") 
 
plt.show() 
```

```python
from math import sqrt
from sklearn.metrics import r2_score 
from sklearn.metrics import mean_squared_error
rsquared=r2_score(y_test, reg.predict(X_test), sample_weight=None)
mean_squared=mean_squared_error(y_test, reg.predict(X_test), sample_weight=None)
print(rsquared)
print(sqrt(mean_squared))
print(mean_squared)
```

### Walk Forward Validation 

#####  Run a continuous loop to train & predict data by appending predicted row from test data to the train data. (Walk Forward Validation)


```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.009,random_state=26)
#reg = linear_model.LinearRegression()
#taking small test_size due to low computational power of muy desktop

# #     # train the model using the training sets

y_test_hat = []
for idx in X_test.index.values:
    # You may use any model of your choice here.
    rf = py.Earth()
    rf.fit(X_train, y_train)

    y_test_hat.extend(rf.predict(X_test.loc[X_test.index == idx]))
    X_train.loc[idx] = X_test.loc[idx]
    y_train.loc[idx] = y_test.loc[idx]
    print(idx)
    
```

```python

y_test_hat = pd.Series(y_test_hat, index=y_test.index)
```

```python
scores = {
    "r2score" : metrics.r2_score(y_test, y_test_hat),
    "rmse" : np.sqrt(metrics.mean_squared_error(y_test, y_test_hat))
}

plt.figure(figsize=(20, 6))
plt.plot(y_test, color = 'r', label = 'y_test actual')
plt.plot(y_test_hat, color = 'b', label = 'y_test predicted')
plt.legend()
#plt.title("Predictions on test set with shuffle={} and label={}".format(SHUFFLE, LABEL))
plt.show()
print(scores)
```

#### Stationary Test

```python
from statsmodels.tsa.stattools import adfuller

def stationarity_test(data):
    dftest = adfuller(data['y1'], autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
    plt.figure(figsize=(16, 7))
    plt.plot(data.index, data['y1'])
    plt.show()

stationarity_test(data)
```

```python
#ACF and PACF plots:
from statsmodels.tsa.stattools import acf, pacf
lag_acf = acf(data['y1'], nlags=20)
lag_pacf = pacf(data['y1'], nlags=20, method='ols')

plt.figure(figsize=(16, 7))
#Plot ACF: 
plt.subplot(121) 
plt.plot(lag_acf, marker="o")
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(data)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(data)),linestyle='--',color='gray')
plt.title('Autocorrelation Function')


#Plot PACF:
plt.subplot(122)
plt.plot(lag_pacf, marker="o")
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(data)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(data)),linestyle='--',color='gray')
plt.title('Partial Autocorrelation Function')
plt.tight_layout()
```
